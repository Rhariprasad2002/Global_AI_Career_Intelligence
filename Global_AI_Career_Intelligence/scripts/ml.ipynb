```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df.columns
```

```python
label_encoder={}

cat_columns=['job_title', 'salary_currency', 'experience_level',
       'employment_type', 'company_location', 'company_size',
       'employee_residence', 'remote_ratio', 'required_skills',
       'education_required', 'industry']
for i in cat_columns:
    le = LabelEncoder()
    df[i] = le.fit_transform(df[i])
    label_encoder[i] = le 
    
```

```python
for i in df:
    sns.boxplot(df[i])
    plt.title(f"Distribution of {i}",color="red")
    plt.show()
```

```python
for i in df:
    sns.histplot(df[i])
    plt.title(f"Distribution of {i}",color="red")
    plt.show()
```

```python
orginal_shape=df.shape

col=df[["salary_currency","salary_usd"]]

for i in col:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    df=df[(df[i]>=lower)&(df[i]<=upper)]

# shape after cleaning
new_shape=df.shape

# Print result
print(f'orginal_shape : {orginal_shape}')
print(f'After removing outliers : {new_shape}')
print(f'Rows removed : {orginal_shape[0]-new_shape[0]}')
```

```python
# After removing outliers

for i in df:
    sns.boxplot(df[i])
    plt.title(f"Distribution of {i}",color="red")
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("x_train_values:",x_train.shape)
print("x_test_values:",x_test.shape)
print("y_train_values:",y_test.shape)
print("y_test_values:",y_test.shape)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
# df_encoded = x.copy()  # make sure this is your encoded DataFrame
# df_encoded.to_csv(r"C:\Global_AI_Career_Intelligence\Data\ai_jobs_dataset_encoded.csv", index=False)
# print("âœ… Encoded dataset saved successfully!")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
import numpy as np

scaled_models = {
    "Linear Regression": LinearRegression(),
    "KNN": KNeighborsRegressor(),
    "SVR": SVR()
}
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

for name, model in scaled_models.items():
    model.fit(x_train_scaled, y_train)
    y_pred = model.predict(x_test_scaled)
    r2= r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)

    print(f"{name:<30} | RÂ²: {r2:.4f} | RMSE: {rmse:.4f} | MAE: {mae:.4f}")
  
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor
from xgboost import XGBRegressor

tree_models={
    "AdaBoostRegressor":AdaBoostRegressor(),
    "GradientBoostingRegressor":GradientBoostingRegressor(),
    "RandomForestRegressor":RandomForestRegressor(),
    "DecisionTreeRegressor":DecisionTreeRegressor()
}
```

```python
for name,model in tree_models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    r2= r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)

    print(f"{name:<30} | RÂ²: {r2:.4f} | RMSE: {rmse:.4f} | MAE: {mae:.4f}")
```

```python
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# âœ… Make predictions
y_pred = gbr.predict(x_test)

# âœ… Evaluate the model
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

# âœ… Print results clearly
print("ðŸ“Š Model Evaluation Metrics")
print("----------------------------")
print(f"RÂ² Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```
